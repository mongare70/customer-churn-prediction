```python
df['LastInteractionDate'] = pd.to_datetime(df['LastInteractionDate'], errors='coerce')
```

```python
df['Frequency'].dtype
```

```python
currentDate = df['LastInteractionDate'].max()

df['recency'] = currentDate - df['LastInteractionDate']
df['frequency'] = df['TotalPurchaseFrequency'] * df['Logins']
df['monetary'] = df['TotalPurchaseFrequency']

r_label = range(5, 0, -1)
f_label = range(1,6)
m_label = range(1,6)

df['recency'] = pd.qcut(df['recency'], q=5, labels=r_label)
df['frequency'] = pd.qcut(df['frequency'], q=5, labels=f_label)
df['monetary'] = pd.qcut(df['monetary'], q=5, labels=m_label)

df[['recency', 'frequency', 'monetary']] = df[['recency', 'frequency', 'monetary']].astype('int32')
df['rfm_value'] = df['recency'] + df['frequency'] + df['monetary']

bins = [3, 6, 9, 12, 15]
labels = ['at_risk', 'need_attention', 'loyal', 'premium']
df['customer_segment'] = pd.cut(df['rfm_value'], bins=bins, labels=labels)
df.drop(['recency', 'frequency', 'monetary', 'rfm_value'], axis=1, inplace=True)
```

```python
nominal_columns = ['Gender', 'TotalInteractionType', 'Frequency'] # One hot encoding
ordinal_columns = ['Segment', 'Plan'] # mean encoding
high_cardinality_columns = ['Location', 'ProductList'] # label encoding
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
ohe = OneHotEncoder(sparse_output=False)

encode_data = df.copy()

encoded_columns = []

for col in nominal_columns:
    transformed_col = ohe.fit_transform(encode_data[[col]])
    transformed_df = pd.DataFrame(transformed_col, columns= [f"{col}_{cat}" for cat in ohe.categories_[0]])
    
    encoded_columns.append(transformed_df)

encode_data = pd.concat([encode_data] + encoded_columns, axis=1)
encode_data.drop(nominal_columns, axis=1, inplace=True)
```

```python
le = LabelEncoder()

for col in ordinal_columns:
    encode_data[col] = le.fit_transform(encode_data[col])
```

```python
for col in high_cardinality_columns:
    encode_data[col] = encode_data.groupby(col)['TotalPurchaseValue'].transform('mean')
```

```python
encode_data.drop(columns='customer_segment', inplace=True)
```

```python
date_col_to_engineer = encode_data.select_dtypes(include='datetime64').columns.to_list()
for col in date_col_to_engineer:
    encode_data[col + '_Year'] = encode_data[col].dt.year
    encode_data[col + '_Month'] = encode_data[col].dt.month
    encode_data[col + '_Day'] = encode_data[col].dt.day

    encode_data.drop(columns = col, inplace=True)
```
