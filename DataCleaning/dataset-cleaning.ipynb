# Feature Engineering

```python
df['PurchaseHistory'].iloc[0]
```

```python
# Add TotalPurchaseValue column
import ast

# If PurchaseHistory is a string column, convert it to list of dicts
df["PurchaseHistory"] = df["PurchaseHistory"].apply(ast.literal_eval)

df["TotalPurchaseValue"] = df["PurchaseHistory"].apply(
    lambda purchases: sum(item["Frequency"] * item["Value"] for item in purchases)
)
```

```python
# Add the SubscriptionPlan from the SubscriptionDetails column
df['SubscriptionDetails'] = df['SubscriptionDetails'].apply(ast.literal_eval)
df['SubscriptionPlan'] = df['SubscriptionDetails'].apply(lambda x: x['Plan'])
```

```python
# Add TotalLatePayments column from the PaymentHistory column
df['PaymentHistory'] = df['PaymentHistory'].apply(ast.literal_eval)
df['TotalLatePayments'] = df['PaymentHistory'].apply(
    lambda LatePayments: sum(item["Late_Payments"] for item in LatePayments)
)
```

```python
# Add PageViews Column from the WebsiteUsage column
df["WebsiteUsage"] = df["WebsiteUsage"].apply(ast.literal_eval)
df["PageViews"] = df["WebsiteUsage"].apply(lambda x: x["PageViews"])
```

```python
# Add TimeSpentOnWebsite Column from the WebsiteUsage column
df["TimeSpentOnWebsite"] = df["WebsiteUsage"].apply(lambda x: x["TimeSpent(minutes)"])
```

```python
# Add NumberOfLogins column from the EngagementMetrics column
df["EngagementMetrics"] = df["EngagementMetrics"].apply(ast.literal_eval)
df["NumberOfLogins"] = df["EngagementMetrics"].apply(lambda x: x["Logins"])
```

```python
# Add FrequencyOfInteractions column from the EngagementMetrics column
df["FrequencyOfInteractions"] = df["EngagementMetrics"].apply(lambda x: x["Frequency"])
```

```python
# Add Rating column from the Feedback column
df["Feedback"] = df["Feedback"].apply(ast.literal_eval)
df["Rating"] = df["Feedback"].apply(lambda x: x["Rating"])
```

```python
# Drop used columns
df = df.drop(columns=['PurchaseHistory', 'SubscriptionDetails', 'PaymentHistory', 'WebsiteUsage', 'EngagementMetrics', 'Feedback'])
```
